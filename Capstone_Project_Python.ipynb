```
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline

df = pd.read_excel('https://github.com/emhy1115/public-school-teacher-salaries/blob/master/public-school-teacher-salaries%20(3).xlsx?raw=true')
#df['Year'] = df[df.columns[1990:]].apply(lambda x:','.join(x.dropna().astype(str)),axis=1)
user_Input = input("What is the fist State you want the average for? (Type United States if you want the average for whole country): ")
user_Input2 = input("What is the second State you want the average for? (Type United States if you want the average for whole country): ")
print("\n")
df_input = df[(df['State'] == user_Input)].round(decimals=2)
print(pd.pivot_table(df_input, index=user_Input).T)
print("\n")
df_input2 = df[(df['State'] == user_Input2)].round(decimals=2)
print(pd.pivot_table(df_input2, index=user_Input2).T)
pd.pivot_table(df_input, index=user_Input).T.plot(figsize=(15,8))
pd.pivot_table(df_input2, index=user_Input2).T.plot(figsize=(15,8))
```
